```python
import pandas as pd
import json
pd.set_option('display.max_rows', 100)
pd.set_option('display.max_columns', 100)
pd.set_option('display.width', 200)
pd.set_option('expand_frame_repr', True)
pd.set_option('display.max_colwidth', 500)
prompts=pd.read_csv("C:/Users/santi/OneDrive - AYUNTAMIENTO DE MURCIA/IA/drones_sa/ROSI.csv", encoding="ISO 8859-1")
training=[]
prompts.at[0,"Prompt"]
```

```python
f = open("C:/Users/santi/OneDrive - AYUNTAMIENTO DE MURCIA/IA/drones_sa/rosi.json", "w")
for i in range(86):
    f.write(json.dumps(
        {
            "messages": [
                {
                    "role": "system",
                    "content": "Eres un chatbot educado",
                },
                {
                    "role": "user",
                    "content": prompts.at[i,"Prompt"] + " END",
                },
                {
                    "role": "assistant",
                    "content": prompts.at[i,"Completion"],
                },
            ]
        }
    ))
    f.write("\n")
    
f.close()
```
